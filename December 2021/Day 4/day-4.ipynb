You're already almost 1.5km (almost a mile) below the surface of the ocean, already so deep that you can't see any sunlight. What you can see, however, is a giant squid that has attached itself to the outside of your submarine.

Maybe it wants to play bingo?

Bingo is played on a set of boards each consisting of a 5x5 grid of numbers. Numbers are chosen at random, and the chosen number is marked on all boards on which it appears. (Numbers may not appear on all boards.) If all numbers in any row or any column of a board are marked, that board wins. (Diagonals don't count.)

The submarine has a bingo subsystem to help passengers (currently, you and the giant squid) pass the time. It automatically generates a random order in which to draw numbers and a random set of boards (your puzzle input).

Visit https://adventofcode.com/2021/day/4 for example

The score of the winning board can now be calculated. Start by finding the sum of all unmarked numbers on that board; in this case, the sum is 188. Then, multiply that sum by the number that was just called when the board won, 24, to get the final score, 188 * 24 = 4512.

To guarantee victory against the giant squid, figure out which board will win first. 

What will your final score be if you choose that board?

```python
# Function to check a board for a bingo
def check_board(marked_board):
    bingo = False;
    # Checking rows
    total = 0
    for row in marked_board:
        for num in row:
            total = total + num
        if(total == len(row)):
            #print(f"{total, len(row)} row")
            bingo = True
            return bingo
        total = 0
    # Checking columns
    # Use this method OR sum every element in a column
    # (if sum == len(marked_board)...Bingo=true)
    total1 = 0
    T = marked_board.T
    for row1 in T:
        for num in row1:
            total1 = total1 + num
        if(total1 == len(row)):
            #print(f"{total, len(row)} row")
            bingo = True
            return bingo 
        total1 = 0
    return bingo
   
```

```python
# Create a function to check the first board that has a bingo
# and return board number
def check_all_boards(marked_boards):
    #print(marked_boards)
    for i, board in enumerate(marked_boards):
        if(check_board(board)):
            return True, i
    return False, None      
```

```python
# Function to change marked_board if number is present in board
def mark_board(board, marked_board, num):
    m = marked_board[:]
    #print(num)
    for i, row in enumerate(board):
        for j, val in enumerate(row):
            if num == val:
                m[i][j] = 1
    return m
```

```python
# Function to mark all boards
def mark_all_boards(boards, marked_boards, num):
    marked = []
    for i, board in enumerate(boards):
        marked.append(mark_board(board, marked_boards[i], num))
    return marked;
```

```python
# Function to find the winning board and the maked numbers
def play_bingo(boards, nums):
    marked = np.zeros(boards.shape, dtype='int')
    for i, num in enumerate(nums):
        marked = mark_all_boards(boards, marked, num)
        isBingo, bingo_board = check_all_boards(marked)
        if(isBingo):
            return boards[bingo_board], marked[bingo_board], num
    return 'No bingo'
```

## --- Part Two ---
On the other hand, it might be wise to try a different strategy: let the giant squid win.

You aren't sure how many bingo boards a giant squid could play at once, so rather than waste time counting its arms, the safe thing to do is to figure out which board will win last and choose that one. That way, no matter which boards it picks, it will win for sure.

In the above example, the second board is the last to win, which happens after 13 is eventually called and its middle column is completely marked. If you were to keep playing until this point, the second board would have a sum of unmarked numbers equal to 148 for a final score of 148 * 13 = 1924.

Figure out which board will win last. Once it wins, what would its final score be?

Here we can rewrite the `play_bingo()` function so it returns a list of the winning boards and their markers in the order at which they won

Or you can mutate a variable every win instead of saving every win in an array

```python
def play_bingo2(boards, nums):
    marked = np.zeros(boards.shape, dtype='int')
    boards_s = boards[:].tolist()
    win_boards = []
    win_markers = []
    win_nums = [] # Numbers that triggered their wins
    for i, num in enumerate(nums):
        marked = mark_all_boards(boards_s, marked, num)
        isBingo, bingo_board = check_all_boards(marked)
        while(isBingo):
            win_boards.append(boards_s[bingo_board])
            win_markers.append(marked[bingo_board])
            win_nums.append(num)
            # Remove winners from boards
            boards_s.pop(bingo_board)
            marked.pop(bingo_board)
            isBingo, bingo_board = check_all_boards(marked)
    return win_boards, win_markers, win_nums
```
